<h2>Linear Congruential Generator (LCG)</h2>
X<sub>n+1</sub> = (a * X<sub>n</sub> + c) % m
<br>

<ul>
<li>X = Current random value</li>
<li>a = multiplier</li>
<li>c = increment</li>
<li>m = modulus</li>
<li>% = modulo (reminder)</li>
</ul>


```python
class LCG:
    def __init__(self, seed):
        self.a = 1664525
        self.c = 10103904223
        self.m = 2**32
        self.state = seed

    def random(self):
        self.state = (self.a * self.state + self.c) % self.m
        return self.state


rng = LCG(seed=1234)
```

```python
print(rng.random())
```
