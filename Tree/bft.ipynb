<h2>Breadth-First Traversal (Level Order)</h2>
<h4>This is not recursion â€” it uses a queue</h4>

When used
<ul>
<li>Shortest path in tree</li>
<li>Minimum depth</li>
<li>Level printing</li>
<li>Tree serialization</li>
</ul>

```python
from collections import deque
def levelOrder(node):
  if not node:
    return None

  queue = deque([node])

  while queue.length:
    node = queue.popleft()

    print(node.val)

    if node.left:
       queue.append(node.left)

    if node.right:
       queue.append(node.right)
```
