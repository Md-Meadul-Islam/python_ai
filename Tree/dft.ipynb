<h2>Depth-First Traversals (DFS)</h2>

<h4>Go deep into a branch before moving sideways</h4>
There are three main DFS traversals
<ol>
<li>In-order Traversal (Left ‚Üí Root ‚Üí Right)</li>
<li>Pre-order Traversal (Root ‚Üí Left ‚Üí Right)</li>
<li>Post-order Traversal (Left ‚Üí Right ‚Üí Root)</li>
</ol>

```python
def dfs(node) :
  if not node:
    return

  # PREORDER

  dfs(node.left)

  # INORDER

  dfs(node.right)

  # POSTORDER
```

<h3>In-order Traversal (Left ‚Üí Root ‚Üí Right)</h3>
When used
<ul>
<li>BST ‚Üí produces sorted values</li>
<li>Validate BST</li>
<li>Kth smallest element</li>
</ul>

```python
def inorder(node):
  if not node:
    return

  inorder(node.left)
  print(node.val)#print or use current node value
  inorder(node.right)
```

<h3>Pre-order Traversal (Root ‚Üí Left ‚Üí Right)</h3>
When used
<ul>
<li>Copying a tree</li>
<li>Serialization <i>üëçconverting a tree into a storable/transmittable format</i></li>
<li>Creating tree structure</li>
<li>Expression trees</li>
</ul>

```python
def preorder(node):
  if not node:
    return None

  print(node.val)#use current node value or print
  preorder(node.left)
  preorder(node.right)
```

<h3>Post-order Traversal (Left ‚Üí Right ‚Üí Root)</h3>
When used
<ul>
<li>Delete tree</li>
<li>Calculate subtree values</li>
<li>Height calculation</li>
<li>Expression evaluation</li>
</ul>

```python
def preorder(node):
  if not node:
    return None

  preorder(node.left)
  preorder(node.right)
  print(node.val)#use current node value or print
```
