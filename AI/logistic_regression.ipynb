<h2>Logistic Regression</h2>
<h3> Sigmoid Function Y = 1 / (1+e<sup>-(a<sub>0</sub>+a<sub>1</sub>X)</sup>)</h3>
a<sub>0</sub> = Intercept(bias)<br>
a<sub>1</sub> = Slop(weight) / how fast probability increases<br>
X = Indepentent Variable<br>
Y = Dependent Variable

```python
import numpy as np

study_hours = np.array([2,3,4,5,6,7,8,9,10])
exam_results = np.array([0,0,0,0,1,1,1,1,1])
a0 = -1.5
a1 = 0.6
# Linear combination
z = a0 + a1 * study_hours
# Sigmoid function
Y_pred = 1 / (1 + np.exp(-z))

print(Y_pred)
```
