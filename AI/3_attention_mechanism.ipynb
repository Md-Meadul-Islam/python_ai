```python
import numpy as np
```

```python
def softmax(x):
    e = np.exp(x)
    return e / np.sum(e)

# Example vectors
Q = np.array([1, 0, 1])
K = np.array([
    [1, 0, 1],
    [0, 1, 0],
    [1, 1, 0]
])
V = np.array([
    [10, 0],
    [0, 10],
    [5, 5]
])

scores = Q @ K.T          # similarity check
weights = softmax(scores) # importance
output = weights @ V # weighted information

print(output)
```
