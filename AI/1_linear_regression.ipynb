<h2>Linear Regression</h2>
Linear Regression predicts a value by taking a weighted sum of input features.

<h3>Y = mX + b</h3>
b = Intercept(bias)<br>
m = Slop(weight) / how fast probability increases<br>
X = Independent Variable<br>
Y = Dependent Variable

```python
import numpy as np
```

```python
# input data (rows = samples, cols = features) or (house size, rooms)
X = np.array([
    [1500, 3],
    [1800, 4],
    [1200, 2]
])
```

```python
# weights (importance of features)
W = np.array([0.5, 1.2])
```

```python
# bias
b = 10

# prediction
y_pred = X @ W + b #price = size*0.5 + rooms*1.2 + bias

print(y_pred)
```

<h2>Pizza Price Predection</h2>
Train Model

```python
#Train Model
diameters = np.array([8,10,12,13,15,17,20,25]) # Independent (X)
prices = np.array([100,130,160,170,180,200,230,300]) # Dependent (Y)

mean_X = np.mean(diameters) # mean/average
mean_Y = np.mean(prices)

deviations_X = diameters - mean_X #differs
deviations_Y = prices - mean_Y #differs

product_of_deviations = np.multiply(deviations_X, deviations_Y)
sum_of_product_of_deviations = np.sum(product_of_deviations)

square_of_deviations_for_X = np.square(deviations_X)
sum_of_square_of_deviations_for_X = np.sum(square_of_deviations_for_X)

m = sum_of_product_of_deviations / sum_of_square_of_deviations_for_X
b = mean_Y - (m*mean_X)
print(m)#slop
print(b)#intercept
```

Now Calculate some prediction

```python
#suppose i made 20 diameters pizza, how much price will be?
prediction = m * 20 + b
print(prediction)
new_diameters = np.array([8,10,12,13,15,17,20,25])#some pizzas diameters
predictions = m * new_diameters + b
print(predictions)
#Visualize Error (Very important practice)
error = prices - predictions
print(error)
```
