<h1>Neural Network (Single Layer)</h1>
A Neural Network is just: Linear Regression + activation function

output = activation(X @ W + b)

```python
import numpy as np
```

```python
def relu(x):#Rectified Linear Unit
    return np.maximum(0, x)

X = np.array([
   [-1500, 3],
   [1800, 4],
   [1200, 2]
])

W = np.array([0.5, 1.2])
b = -1

z = X @ W + b       # linear part
output = relu(z)   # non-linearity

print(output)
```

<h3>Train Model</h3>

```python
#Train Model
diameters = np.array([8,10,12,13,15,17,20,25]) # Independent (X)
prices = np.array([100,130,160,170,180,200,230,300]) # Dependent (Y)

mean_X = np.mean(diameters) # mean/average
mean_Y = np.mean(prices)

deviations_X = diameters - mean_X #differs
deviations_Y = prices - mean_Y #differs

product_of_deviations = np.multiply(deviations_X, deviations_Y)
sum_of_product_of_deviations = np.sum(product_of_deviations)

square_of_deviations_for_X = np.square(deviations_X)
sum_of_square_of_deviations_for_X = np.sum(square_of_deviations_for_X)

m = sum_of_product_of_deviations / sum_of_square_of_deviations_for_X #m = covariance(X, Y) / variance(X)
b = mean_Y - (m*mean_X)#b = mean_Y − m·mean_X
print(m)#slop
print(b)#intercept
```

```python
def predict(X):
    return m * X + b
```

```python
#suppose i made 20 diameters pizza, how much price will be?
print(predict(20))
```

So you’ve already built the training step for simple linear regression.

Step 2 — Predict Multiple Values (NumPy style)

```python
new_diameters = np.array([8,10,12,13,15,17,20,25])#some pizzas diameters
new_predictions = predict(new_diameters)
print(new_predictions)
```

<h4>Step 3 — Visualize Error (Very important practice)</h4>

```python
error = prices - new_predictions
print(error)#This shows how wrong the model is.
```

<h4>Step 4 — Turn Your Model into Neural-Network Form</h4>
<h3>Y = X@W + b</h3>
b = Intercept(bias)<br>
X = Slop(weight) / how fast probability increases<br>
W = Independent Variable<br>

```python
W = np.array([m])#make slop into array
b = b
X = diameters.reshape(-1, 1)#row->column
print(diameters, X)
y_pred = X @ W + b#[8m+b, 10m+b, 12m+b]
print(y_pred)
```
