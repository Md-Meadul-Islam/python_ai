<h2>Transformer (Tiny Mental Model)</h2>
A Transformer is: Attention + Feedforward Neural Network
<br>
Flow: X → Attention → Linear → Activation

```python
import numpy as np

def softmax(x):
    e = np.exp(x)
    return e / np.sum(e, axis=-1, keepdims=True)

X = np.array([
    [1, 0],
    [0, 1],
    [1, 1]
])

# Attention weights
Wq = np.array([[1, 0], [0, 1]])
Wk = np.array([[1, 0], [0, 1]])
Wv = np.array([[1, 2], [3, 4]])

Q = X @ Wq
K = X @ Wk
V = X @ Wv

scores = Q @ K.T
weights = softmax(scores)

attention_out = weights @ V

print(attention_out)
```
